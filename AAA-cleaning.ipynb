#Cleaning a dataset of various clothes and accessories using pandas

```python
import pandas as pd
import numpy as np
```

```python
datas = pd.read_csv('https://raw.githubusercontent.com/BriceNW/UE_AAA/main/FashionDataset.csv')
```

```python
datas
```

```python
datas.drop([datas.columns[0]],axis=1,inplace=True)

columns_names = datas.columns.tolist()
```

```python
#Renaming the columns:
new_names = ['Marque', 'Description', 'Tailles', 'PrixInit', 'PrixSoldé', 'Réduc', 'Catégorie']

for i in range(len(columns_names)):
  datas.rename(columns = {columns_names[i] : new_names[i]}, inplace=True)
```

```python
#Cleaning useless informations:
datas['PrixInit'].replace('Rs\n','', regex=True, inplace=True)
datas['Tailles'].replace('Size:','', regex=True, inplace=True)
datas['Réduc']= datas['Réduc'].str.replace('% off','')
```

```python
#Splitting categories and genders into 2 columns:
datas[['Catégorie','Genre']] = datas['Catégorie'].str.split('-', expand=True)
```

```python
datas.replace('Nan',np.nan, regex=True, inplace=True)
```

```python
#Replacing 'Nan' brands by the brand most present in the dataset:
nan_marque = datas.Marque.value_counts().index.tolist()[0]

datas.Marque.fillna(nan_marque, inplace=True)
```

```python
nan_prixin = datas.PrixInit.astype(float).mean()
nan_prixso = datas.PrixSoldé.astype(float).mean()
nan_red = datas.Réduc.astype(float).mean()
```

```python
#Replacing every nan by the value of each category that is the most present:
nan_des = datas.Description.value_counts().index.tolist()[0]
nan_taille = datas.Tailles.value_counts().index.tolist()[0]
nan_cat = datas.Catégorie.value_counts().index.tolist()[0]
nan_g = datas.Genre.value_counts().index.tolist()[0]

datas.Description.fillna(nan_des, inplace=True)
datas.Tailles.fillna(nan_taille, inplace=True)
datas.PrixInit.fillna(nan_prixin, inplace=True)
datas.PrixSoldé.fillna(nan_prixso, inplace=True)
datas.Réduc.fillna(nan_red, inplace=True)
datas.Catégorie.fillna(nan_cat, inplace=True)
datas.Genre.fillna(nan_g, inplace=True)
```
